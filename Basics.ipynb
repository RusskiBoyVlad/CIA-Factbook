```python
%%sql
select *
from facts
limit 5
```

```python
%%sql

select MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
from facts
```

```python
%%sql

select name
from facts 
where population == (select min(population) from facts);
```

```python
%%sql

select name
from facts 
where population == (select max(population) from facts);
```

```python
%%sql

select MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
from facts
where name != "World"
```

```python

%%sql
select population, area
from facts
```

```python
%%sql
select population, area
from facts
where population >= (select avg(population) 
                     from facts where name <> "World")
                    and area < (select avg(area) from facts where name <> "World");
```

```python
%%sql

select name, population
from facts
where population == (select Max(population) from facts where name <> "World")
```

```python
%%sql

select name, population_growth
from facts
where population_growth == (select Max(population_growth) from facts where name <> "World")
```

```python
%%sql
select *
from facts
limit 1
```

```python
%%sql

select name as country, cast(area_land as float)/area_water as "water_t_land"
from facts
order by water_t_land Desc
limit 5
```

```python
%%sql

select name, MAX((population) *(1+population_growth/100)) as pop_next_year
from facts
where name <> "World"        
```
